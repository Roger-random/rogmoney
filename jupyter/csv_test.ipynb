```python
import csv
import datetime

csv_list = ('./data/2016.csv', 
            './data/2017.csv',
            './data/2018.csv',
            './data/2019.csv',
            './data/2020.csv',
            './data/2021.csv',
            './data/2022.csv',
            './data/2023.csv',
            './data/2024.csv')

vendors = list()
categories = list()
parsed_list = list()

for filename in csv_list:
    with open(filename, newline='') as csvfile:
        cd = csv.DictReader(csvfile)
        for row in cd:
            parsed = {}
            for column, data in row.items():
                data = data.strip()
                if column.endswith('Date'):
                    parsed['Date'] = data
                elif column == 'Vendor':
                    if data not in vendors:
                        vendors.append(data)
                    parsed['Vendor'] = data
                elif len(data)>0:
                    if column not in categories:
                        categories.append(column)
                    parsed[column]=float(data)
            parsed_list.append(parsed)
    print("{} items to {} vendors in {} categories after adding {}".format(len(parsed_list), len(vendors), len(categories), filename))
```
